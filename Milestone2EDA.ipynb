# Highest Percentage of Foreign players in a domestic league

Let's see which club has the highest percentage of foreign players

```python
clubs.sort_values(by='foreigners_percentage',ascending=False).head(5)
```

All the players in the team AS Monaco FC are foreigners. Which is expected as it is based in Monaco which is one of the smallest country with just one municipality

Now we group by the different national leagues and see which league has the highest percentage of foreign players.

```python
#first find first tier domestic leagues from the competitions table
leagueCompetitions = competitions[(competitions['type']=='domestic_league')&(competitions['sub_type']=='first_tier')]
#Left join clubs and league competitions
merged_club_competition = clubs.merge(leagueCompetitions, left_on='domestic_competition_id', right_on='domestic_league_code') #We have merged competition and club

foreigners_details = merged_club_competition.groupby('name_y')['foreigners_percentage'].mean().reset_index()
foreigners_details.rename(columns={'name_y': 'League Name', 'foreigners_percentage': 'Percentage of Foreigners'}, inplace=True)
foreigners_details_sorted=foreigners_details.sort_values(by='Percentage of Foreigners', ascending=False) #Sort the data by percentage in a descending order
print(foreigners_details_sorted)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Set the style of the visualization
sns.set(style='whitegrid')

# Create a bar plot with updated parameters
plt.figure(figsize=(12, 6))
sns.barplot(data=foreigners_details_sorted, x='League Name',
            y='Percentage of Foreigners', hue='League Name',
            palette='viridis', legend=False)

# Rotate x labels for better readability
plt.xticks(rotation=45)

# Add title and labels
plt.title('Average Percentage of Foreign Players by League')
plt.xlabel('League Name')
plt.ylabel('Percentage of Foreigners')

# Show the plot
plt.tight_layout()
plt.show()

```

The Premier League which is the league with the highest percentage of foreign players, followed by jupiler pro league and then the scottish premiership

# Player Position vs Red Cards

**Now that we have data on how many red cards a player has. We can find out the relation between player position and red cards**

```python
# Filter out rows where position is None or empty
filtered_lineups = game_lineups[game_lineups['position'].notna() & (game_lineups['position'] != '') & (game_lineups['position']!='Attack') & (game_lineups['position']!='midfield')
                   & (game_lineups['position']!='Defender')]

#Group by player_id and get the most common position
def most_common_position(x):
    if x.empty:
        return None
    return x.value_counts().index[0]

position_data = filtered_lineups.groupby('player_id')['position'].agg(most_common_position).reset_index()
position_data.columns = ['player_id', 'position']
print(position_data)
```

```python
merged_position_red_cards = Total_red_cards.merge(position_data,left_on='player_id',right_on='player_id')
red_cards_by_position = merged_position_red_cards.groupby('position')['red_cards'].sum().reset_index()
red_cards_by_position = red_cards_by_position.sort_values(by='red_cards', ascending=False)
#print(red_cards_by_position)
```

```python
#plot pie chart to show red card x position
plt.figure(figsize=(12, 6))

plt.pie(red_cards_by_position['red_cards'], labels=red_cards_by_position['position'], autopct='%1.1f%%')

plt.title('Percentage of red cards by position', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

We find out that centre backs are the ones who get the most red cards than other positions

# Analysis on player market value

```python
#First we merge players and player valuations
player_value = player_valuations.merge(players,left_on='player_id',right_on='player_id')
#we will check the market value for the player Vinicius Junior
vinicius_junior_value = player_value[player_value['name'].astype(str).str.contains('Vinicius Junior',case=False)]
```

```python
#plot linechart to show change in player valuation
plt.figure(figsize=(12, 6))

sns.lineplot(data=vinicius_junior_value, x='date', y='market_value_in_eur_x', marker='o', color='blue')

plt.title('Changing Market Valuation of Vinicius Junior', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Market Value (in millions)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Let's find out which player position has the highest Average market value

```python
game_lineups.head()
```

```python
merged_data = pd.merge(players, position_data[['player_id', 'position']], on='player_id', how='left')
mean_market_values = merged_data.groupby('position_y')['market_value_in_eur'].mean().reset_index()
merged_data.head()
```

```python

plt.figure(figsize=(12, 8))

sns.barplot(data=mean_market_values, x='position_y', y='market_value_in_eur', palette='Blues_d')
plt.title('Player Position vs. Market Value', fontsize=16)
plt.xlabel('Player Position', fontsize=12)
plt.ylabel('Market Value (M€)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Player Position')
plt.tight_layout()

plt.show()
```

Hence we find out that attacking midfielders, strikers and wingers as an average have a higher market value than other positions. While goalkeepers on average have the lowest
